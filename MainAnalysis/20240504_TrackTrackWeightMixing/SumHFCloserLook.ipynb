```python
import ROOT
```

```python
FileSxB1 = ROOT.TFile("PbPbDataxMinBiasData_Centrality030_Kaya.root")
FileSxB2 = ROOT.TFile("PbPbDataxMinBiasData2_Centrality030_Kaya.root")
FileWeight = ROOT.TFile("DataWeight_Centrality030.root")
```

```python
Canvas = ROOT.TCanvas("Canvas", "")
```

```python
FileSxB1.ls()
```

```python
HSxB1_SignalHF  = FileSxB1.Get("HSignalHFLinaer")
HSxB1_MatchedHF = FileSxB1.Get("HMatchedHFLinear")
HSxB2_SignalHF  = FileSxB2.Get("HSignalHFLinaer")
HSxB2_MatchedHF = FileSxB2.Get("HMatchedHFLinear")
HWeight = FileWeight.Get("HRatio")
HTarget = FileWeight.Get("HDY")
```

```python
HSxB1_MatchedHF.SetStats(0)

HSxB1_MatchedHF.SetLineWidth(2)
HSxB1_MatchedHF.SetLineColor(ROOT.kBlue)
HSxB1_MatchedHF.SetMarkerColor(ROOT.kBlue)
HSxB2_MatchedHF.SetLineWidth(2)
HSxB2_MatchedHF.SetLineColor(ROOT.kRed)
HSxB2_MatchedHF.SetMarkerColor(ROOT.kRed)

HSxB1_MatchedHF.Draw()
HSxB2_MatchedHF.Draw("same")
Canvas.Draw()
```

```python
HSxB1_MatchedHFRatio = HSxB1_MatchedHF.Clone("HSxB1Ratio")
HSxB2_MatchedHFRatio = HSxB2_MatchedHF.Clone("HSxB2Ratio")

HSxB1_MatchedHFRatio.Divide(HTarget)
HSxB2_MatchedHFRatio.Divide(HTarget)

HSxB1_MatchedHFRatio.SetLineColor(ROOT.kBlue)
HSxB2_MatchedHFRatio.SetLineColor(ROOT.kRed)

HSxB1_MatchedHFRatio.Draw("hist")
HSxB2_MatchedHFRatio.Draw("hist same")
Canvas.SetLogy(False)
Canvas.Draw()
```

```python
HSxB1_MatchedHFRatio.GetXaxis().SetRangeUser(100000, 120000)
HSxB1_MatchedHFRatio.Draw("hist")
HSxB2_MatchedHFRatio.Draw("hist same")
Canvas.SetLogy(False)
Canvas.Draw()
```
